```python
import pandas as pd
from statsmodels.formula.api import ols
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('dataset.xlsx')
```

# EDA



```python
df.columns = [x.lower().strip().replace(' ','_') for x in df.columns]

print(df.columns)
```

```python
dfPositive = df[df['sars-cov-2_exam_result']=='positive']
dfNegative = df[df['sars-cov-2_exam_result']=='negative']

```

```python
print(f"There are {df['patient_id'].nunique()} patients")
print(f"There are {dfPositive['patient_id'].nunique()} positive patients ")
print(f"There are {dfNegative['patient_id'].nunique()} negative patients ")
```
